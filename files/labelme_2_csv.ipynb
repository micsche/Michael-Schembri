```python
from lxml import etree
from os import listdir
from os.path import isfile, join
from io import StringIO, BytesIO
```

```python
def parseXML(xmlFile):
    context = etree.iterparse(xmlFile)
    csvlist=[""]*6
    pts=0
    for action, elem in context:
        
        if elem.tag == "filename":
            csvlist[0]=elem.text
        
        if elem.tag == "x":
            if pts==0:
                csvlist[1]=elem.text
            else:
                csvlist[3]=elem.text
                
        if elem.tag == "y":
            if pts==0:
                csvlist[2]=elem.text
                pts = pts +1
            else:
                csvlist[4]=elem.text
                print(csvlist)
                pts = 0
        
        if elem.tag == "deleted":
            if elem.text=="1":
                csvlist[5]="deleted"
                
                
        if elem.tag == "name":
            if elem.text != "litter":
                print("ERROR NOT LITTER",csvlist[0])
```

```python
import glob
onlyfiles = glob.glob("./*.xml")
```

```python
for xmlfiles in onlyfiles:
    parseXML(xmlfiles)
```
