```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
```

```python
from os import listdir
from os.path import isfile, join

mypath="/var/www/html/LabelMe/Images/upairtest/"
```

```python
def hist_dataset(mypath):
    onlyfiles = [join(mypath,f) for f in listdir(mypath) if isfile(join(mypath, f))]
    
    r1 = mpimg.imread(onlyfiles[0]).astype(np.float32)

    for jpg in onlyfiles:
        try:
            r1 = r1 + mpimg.imread(jpg)
        except:
            print(jpg)
        
    div=np.max(r1[:,:,:])/255

    red_ch = r1[:, :, 0]/div
    red_ch = red_ch.astype(np.uint8)
    gre_ch = r1[:, :, 1]/div
    gre_ch = gre_ch.astype(np.uint8)
    blu_ch = r1[:, :, 2]/div
    blu_ch = blu_ch.astype(np.uint8)

    hist = cv2.calcHist([red_ch],[0],None,[256],[0,256])
    plt.hist(red_ch.ravel(),256,[0,256])
    plt.title('Histogram for Red scale picture')
    plt.show()

    hist = cv2.calcHist([gre_ch],[0],None,[256],[0,256])
    plt.hist(gre_ch.ravel(),256,[0,256])
    plt.title('Histogram for Green scale picture')
    plt.show()

    hist = cv2.calcHist([blu_ch],[0],None,[256],[0,256])
    plt.hist(blu_ch.ravel(),256,[0,256])
    plt.title('Histogram for Blue scale picture')
    plt.show()
    
    t1=(r1/div).astype(np.uint8)
    plt.imshow(t1)
```

```python
hist_dataset("/var/www/html/LabelMe/Images/upairtest/")
```
